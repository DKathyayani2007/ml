```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20 , 25 , 35 , 55]

plt.plot(x, y)
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['thur' , 'fri' , 'sat' , 'sun']
y = [170,120,250,190] 

plt.bar(x, y)
plt.title('Bar Graph')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.show()
```

```python

import pandas as pd

cars = ['audi' , 'bmw' , 'ford' , 'honda' , 'toyota']
speed = [200, 240, 150, 180, 220]

plt.pie(speed, labels=cars, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
data = [[10,12,14,15,18,20,22], [5,7,9,11,13,15,17], [2,4,6,8,10,12,14]]

plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.show()
```

```python
import numpy as np

np.random.seed(0)

data = np.random.rand(10,10)

plt.imshow(data,cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
