```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={
    'Name':['Alice','Bob','Charlie','David','Eve'],
    'Age':[25,30,35,40,45],
}
df=pd.DataFrame(data)
sns.lineplot(x=df.index,y='Age',data=df)
plt.title('Age of individuals')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()
```

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=iris.target
df['species']=df['species'].map({0:'setosa',1:'versicolor',2:'virginica'})
sns.kdeplot(data=df, x='sepal length (cm)', hue='species', fill=True)
plt.legend()
plt.title('KDE plot of sepal length')
plt.xlabel('Sepal length(cm)')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
iris=sns.load_dataset('iris')
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
plt.title('Scatter plot of sepal length for virginica')
plt.xlabel('Sepal length(cm)')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
iris=sns.load_dataset('iris')
sns.boxplot(x='species',y='petal_length',data=sns.load_dataset('iris'))
plt.title('Box plot of petal length by species')
plt.xlabel('Species')
plt.ylabel('petal length(cm)')
plt.show()
```
