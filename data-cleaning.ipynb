```python
import pandas as pd
import numpy as np

# Creating a deliberately dirty dataset
data = {
    'House_ID': [101, 102, 103, 104, 105, 102, 106, 107], # 102 is a duplicate
    'Price': [250000, 300000, np.nan, '400000', 350000, 300000, 500000, np.nan], # Strings mixed with numbers, NaNs
    'Bedrooms': [3, 4, 3, 5, np.nan, 4, 3, 2], # Missing value
    'Area_SqFt': ['1500', '2000', 'Unknown', '2500', '2200', '2000', '3000', '1200'] # 'Unknown' string causing type issues
}

df = pd.DataFrame(data)
print("--- RAW DATA ---")
print(df)
```

```python
print("\n--- DIAGNOSIS ---")
# 1. Check data types
print("Data Types:\n", df.dtypes)

# 2. Check for missing values
print("\nMissing Values Count:\n", df.isnull().sum())

# 3. Check for exact duplicates
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
print("\n--- CLEANING TYPES ---")
# Convert 'Price' to numeric. 
# We know '400000' is a string. pd.to_numeric fixes this.
df['Price'] = pd.to_numeric(df['Price'])

# Convert 'Area_SqFt' to numeric. 
# The string 'Unknown' will cause an error. 'coerce' forces invalid parsing to become NaN.
df['Area_SqFt'] = pd.to_numeric(df['Area_SqFt'], errors='coerce')

print(df.dtypes)
```

```python
print("\n--- HANDLING DUPLICATES ---")
# Keep the first instance, drop the rest.
df = df.drop_duplicates(keep='first')

# If House_ID is a unique identifier, check for duplicates just in that column
# df = df.drop_duplicates(subset=['House_ID'], keep='first') 

print(df)
```

```python
print("\n--- HANDLING MISSING VALUES ---")

# Fill missing Price with the median
price_median = df['Price'].median()
df['Price'] = df['Price'].fillna(price_median)

# Fill missing Area with the mean (assuming normal distribution for this example)
area_mean = df['Area_SqFt'].mean()
df['Area_SqFt'] = df['Area_SqFt'].fillna(area_mean)

# Fill missing Bedrooms with the median (keeps it as a whole number)
bedroom_median = df['Bedrooms'].median()
df['Bedrooms'] = df['Bedrooms'].fillna(bedroom_median)

print("\n--- FINAL CLEANED DATA ---")
print(df)
print("\nFinal Data Types:\n", df.dtypes)
print("\nFinal Missing Values:\n", df.isnull().sum())
```
